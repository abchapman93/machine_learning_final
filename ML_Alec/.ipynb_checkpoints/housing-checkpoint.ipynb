# Break up into training and test sets

```python
import numpy as np
import numpy.random as rnd
```

```python
#split up the batch into random test sets
def split_train_test(data, test_ratio):
    shuffled_indices = rnd.permutation(len(data))
    test_set_size = int(len(data)*test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]
```

```python
train_set, test_set = split_train_test(housing, 0.2)
```

```python
print(len(train_set), "train +", len(test_set),"test")
```

```python
import hashlib
```

```python
#use each instance's identifier to decide whether or not it should
#go in the test set
#compute a hash of each instance's identifier, keep last byte
#put the instance in the set if it is <= 256 (20% of 256)
def test_set_check(identifier, test_ratio, hash):
    return hash(identifier).digest()[-1] < 256*test_ratio

def split_train_test_by_id(data, test_ratio, id_column, hash=hashlib.md5):
    ids = data[id_column]
    in_test_set = ids.apply(lambda id_: test_set_check(id_, test_ratio, hash))
    return data.loc[~in_test_set], data.loc[in_test_set]
```

```python
#create an income category attribute
#divide the median income by 1.5 to limit number of income categories
#round up using ceil to have discrete categories
#merge all categories greather than 5 into category 5
housing["income_cat"]= np.ceil(housing["median_income"]/1.5)
housing["income_cat"].where(housing["income_cat"]<5, 5.0, inplace=True)
```

```python
#stratify sampling to avoid bias
from sklearn.cross_validation import StratifiedShuffleSplit
```

```python
#stratify based on income category
#create a StratifiedShuffleSplit class

split = StratifiedShuffleSplit(housing["income_cat"], test_size=0.2)
print(split)
```

```python
#choose the indices for each set
train_indices, test_indices = next(iter(split))
print(len(train_indices), len(test_indices))
```

```python
#create the sets using those indices
strat_train_set = housing.loc[train_indices]
strat_test_set = housing.loc[test_indices]
strat_test_set.info()
```

```python
#look at the income category proportions in the full housing dataset
#almost identical to those in the full dataset, not skewed like the random sampling
housing["income_cat"].value_counts()/len(housing)
```

```python
#remove the income_cat attribute 
#so the datais back to its original state
for set in (strat_train_set, strat_test_set):
    set.drop(["income_cat"], axis=1, inplace=True)
```
